```python
# A1 in Rq^{n*k1}; A2 in Rq^{n*k2}
# r in Rq^k1; x in Rq^k2
# y1 $ D_sigma1^k1; y2 $ D_sigma2^k2
# c in Rq
# N = 1024 (bits = 10)
# q = 2^32
# n = 1; k1 = 3; k2 = 3 in Lyu18
# sigma = 27000
from sage.stats.distributions.discrete_gaussian_polynomial import DiscreteGaussianDistributionPolynomialSampler
```

```python
def H(RP, m):
    return RP(m[0].list())
```

```python
# ref: Quantum resistant key-exposure free chameleon hash and applications in redactable blockchain
# Chunhui Wu
class chameleon_hash:
    def __init__(self, N, sigma):
        self.m = 16 # order
        self.n = 1 # rows
        self.sigma = sigma
        self.pp = random_prime(N) # prime
        Z2p = IntegerModRing(2*self.pp)
        PP.<p> = Z2p[]
        self.RP = QuotientRing(PP, PP.ideal(p^self.m+1), names='p')
        
    def Gen(self):
        x_coef = [randint(0,1) for _ in range(self.m)]
        a_coef = [(self.pp*(2*randint(0,self.pp)+1)) for _ in range(self.m)]
        if (vector(a_coef)*vector(x_coef)) % (2*self.pp) == 0:
            x_coef[0] += 1
        X = self.RP(x_coef)
        A = self.RP(a_coef)
        rho = [randint(1, 2*self.pp), DiscreteGaussianDistributionPolynomialSampler(self.RP, self.m, self.sigma)()]
        return A, rho, X
    
    def CH(self, A, u, rho):
        #s = DiscreteGaussianDistributionPolynomialSampler(self.RQ, self.m, self.sigma)()
        #c = randint(1, 2*self.q)
        # rho = [y, s]
        # let c = u
        c = H(self.RP, u)
        return rho[0] + A*rho[1] + self.pp*c
    
    def UF(self, A, X, h, u, rho, u_):
        k = DiscreteGaussianDistributionPolynomialSampler(self.RP, self.m, self.sigma)()
        y_ = h + A*k
        b = randint(0,1)
        s_ = -k + (-1)^b*X*u_
        return [y_, s_]
        
```

```python
class signature_oracle:
    def __init__(self, Rq, N, sigma):
        self.h1 = [DiscreteGaussianDistributionPolynomialSampler(Rq, N, sigma)() for _ in range(3)]
        self.h2 = Rq(list(randrange(-1, 2) for _ in range(N)))
        
    def Sign(self, m):
        return (m, vector(self.h2*vector(self.h1)))
    
    def Verify(self, m, sigma):
        return (sigma==self.Sign(m))
```

```python
# has secret r, x
class Prover:
    def __init__(self, Rq, A1, A2, k1, k2, sigma, N):
        self.honest = True
        self.Rq = Rq
        self.r = []
        self.x = []
        self.A1 = A1
        self.A2 = A2
        self.y1 = [DiscreteGaussianDistributionPolynomialSampler(Rq, N, sigma)() for _ in range(k1)]
        self.y2 = [DiscreteGaussianDistributionPolynomialSampler(Rq, N, sigma)() for _ in range(k2)]
        
    def Commit(self, S, p_set):
        w_ = matrix(self.A1)*vector(self.y1)+matrix(self.A2)*vector(self.y2)
        A = p_set[0]
        CHF = p_set[1]
        rho = p_set[2]
        w = CHF.CH(A, w_, rho)
        return (w_, w, S.Sign(w))
    
    def Prove(self, S, c):
        z1 = vector(self.y1) + c*vector(self.r)
        z2 = vector(self.y2) + c*vector(self.x)
        return (z1, z2, S.Sign(z1), S.Sign(z2))
```

```python
def SQRT(z):
    temp = 0
    for i in z:
        temp += i^2
    return int(temp)^0.5
```

```python
class Verifier:
    def __init__(self, Rq, A1, A2, k1, k2, sigma, N):
        self.Rq = Rq
        self.N = N
        self.sigma = sigma
        self.A1 = A1
        self.A2 = A2
        self.X = 0
        self.CHF = chameleon_hash(N, sigma)
    
    def KeyGen(self):
        return self.CHF.Gen()
    
    def Challenge(self):
        c = DiscreteGaussianDistributionPolynomialSampler(self.Rq, self.N, self.sigma)()
        return c
    
    def Verify(self, S, A, rho, t, w, c, z1, z2, sigma_w, sigma_z1, sigma_z2):
        v1 = (sigma_w == S.Sign(w))
        v2 = (sigma_z1 == S.Sign(z1))
        v3 = (sigma_z2 == S.Sign(z2))
        wh_ = matrix(self.A1)*vector(z1)+matrix(self.A2)*vector(z2)-c*vector(t)
        wh = self.CHF.CH(A, wh_, rho)
        v4 = (SQRT(z1[0].list())<2*self.sigma*self.N^0.5)
        v5 = (SQRT(z2[0].list())<2*self.sigma*self.N^0.5)
        v6 = (wh == w)
        return (v1 and v2 and v3 and v4 and v5 and v6)
```

```python
import time
class NTQRZKP_Protocol:
    def __init__(self, n, k1, k2, seed):
        set_random_seed(seed)
        current_randstate().set_seed_gp()
        
        q = 2^32
        N = 1024
        sigma = 27000
        Zq = Integers(q)
        P.<a> = Zq[]
        Rq = QuotientRing(P, P.ideal(a^N+1), names='a')
        
        A1_ = [[DiscreteGaussianDistributionPolynomialSampler(Rq, N, sigma)() for _ in range(k1-n)] for __ in range(n)]
        I_n = identity_matrix(n)
        A1 = [[] for _ in range(n)]
        for i in range(n):
            for j in range(n):
                A1[i].append(I_n[i][j])
            for k in range(k1-n):
                A1[i].append(A1_[i][k])

        A2_ = [[DiscreteGaussianDistributionPolynomialSampler(Rq, N, sigma)() for _ in range(k2-n-n)] for __ in range(n)]
        Zero_n = matrix(n)
        A2 = [[] for _ in range(n)]
        for i in range(n):
            for z in range(n):
                A2[i].append(Zero_n[i][z])
            for j in range(n):
                A2[i].append(I_n[i][j])
            for k in range(k2-n-n):
                A2[i].append(A2_[i][k])
        # knowledge r, x
        r = [Rq(list(randrange(-1, 2) for _ in range(N))) for _ in range(k1)]
        x = [Rq(list(randrange(-1, 2) for _ in range(N))) for _ in range(k2)]
        t = matrix(A1)*vector(r) + matrix(A2)*vector(x)
        # t = A1*r + A2*x
        
        S = signature_oracle(Rq, N, sigma)
        #========================protocol start=========================
        prover = Prover(Rq, A1, A2, k1, k2, sigma, N)
        verifier = Verifier(Rq, A1, A2, k1, k2, sigma, N)
        
        if prover.honest:
            prover.r = r
            prover.x = x
        else:
            r_ = [DiscreteGaussianDistributionPolynomialSampler(R2, N, sigma)() for _ in range(k1)]
            x_ = [DiscreteGaussianDistributionPolynomialSampler(R2, N, sigma)() for _ in range(k2)]
            rf = []
            xf = []
            for i in range(k1):
                rf.append(Rq(r_[i].list()))
            for i in range(k2):
                xf.append(Rq(x_[i].list()))
            prover.r = rf
            prover.x = xf
        
        start = time.time()
        A, rho, verifier.X = verifier.KeyGen()
        end = time.time()
        self.keygen_time = end - start
        
        p_set = [A, verifier.CHF, rho]
        
        start = time.time()
        w_, w, sigma_w = prover.Commit(S, p_set)
        end = time.time()
        self.commit_time = end - start
        
        start = time.time()
        c = verifier.Challenge()
        end = time.time()
        self.challenge_time = end - start
        
        start = time.time()
        z1, z2, sigma_z1, sigma_z2 = prover.Prove(S,c)
        end = time.time()
        self.prove_time = end - start
        
        start = time.time()
        self.result = verifier.Verify(S, A, rho, t, w, c, z1, z2, sigma_w, sigma_z1, sigma_z2)
        end = time.time()
        self.verify_time = end - start
```

```python
n, k1, k2 = 12, 30, 30
ctr = 0
iteration = 20
total_keygen_time = 0
total_commit_time = 0
total_challenge_time = 0
total_prove_time = 0
total_verify_time = 0

while ctr < iteration:
    seed = time.time_ns()
    print("iteration : %d / %d" % (ctr+1, iteration), end="\r")

    ntqrzkp = NTQRZKP_Protocol(n, k1, k2, seed)

    if ntqrzkp.result:
        ctr += 1
        total_keygen_time += ntqrzkp.keygen_time*10^6
        total_commit_time += ntqrzkp.commit_time*10^6
        total_challenge_time += ntqrzkp.challenge_time*10^6
        total_prove_time += ntqrzkp.prove_time*10^6
        total_verify_time += ntqrzkp.verify_time*10^6
    else:
        pass

print("\n-----------------------------")
print("Total keygen time / Average keygen time       : %.2fus / %.2fus" % (total_keygen_time, total_keygen_time/iteration))
print("Total commit time / Average commit time       : %.2fus / %.2fus" % (total_commit_time, total_commit_time/iteration))
print("Total challenge time / Average challenge time : %.2fus / %.2fus" % (total_challenge_time, total_challenge_time/iteration))
print("Total prove time / Average prove time         : %.2fus / %.2fus" % (total_prove_time, total_prove_time/iteration))
print("Total verify time / Average verify time       : %.2fus / %.2fus" % (total_verify_time, total_verify_time/iteration))
```
