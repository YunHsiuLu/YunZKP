```python
x = np.array([1,2,3])
szkp = np.array([[0.18, 0.01+0.09, 0.45], [1.17, 0.02+0.57, 2.85], [7.93, 0.02+3.89, 19.38]])
bdvp = np.array([[0.44, 0.01+0.09, 0.68], [2.79, 0.01+0.56, 4.50], [19.32, 0.03+3.88, 30.77]])
dvp  = np.array([[0.20+0.10, 0.19, 0.18], [1.20+0.61, 1.35, 1.21], [7.90+3.93, 7.88, 7.82]])
fig, ax = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Performance: SZKP [19] v.s. DVP [8] v.s. BDVP', fontsize=16)
titles = ['512 bits', '1024 bits', '2048 bits']
labels = ['SZKP [19]', 'DVP [8]', 'BDVP']
handles = [0,0,0]
for i in range(3):
    handles[0] = ax[i].bar(x-0.3, szkp[i], 0.3, label=labels[0])
    handles[1] = ax[i].bar(x, dvp[i], 0.3, label=labels[1])
    handles[2] = ax[i].bar(x+0.3, bdvp[i], 0.3, label=labels[2])
    ax[i].set_ylabel('CPU times (ms)', fontsize=13)
    ax[i].set_title(titles[i], fontsize=16)
    ax[i].set_xticks(x, ['Commit', 'Challenge+\nProve', 'Verify'], rotation=30, fontsize=13)

fig.legend(handles, labels, fontsize=16)
#plt.show()
fig.savefig('SZKPvsDVPvsBDVP.png')
```
